# loc and iloc

Both are used for data selection
## iloc
uses the numeric index of row and column (eg 4th row and 5th column)

## loc
uses actual name of the row and column (e.g. 4th row in '*address*' column or '*Alice*' row in '*address*' column)

```python
df_csv.iloc[5]
```

```python
df_csv.iloc[5,2]
```

```python
df_csv.iloc[5,:2]
```

```python
df_csv.iloc[:2]
```

```python
df_csv.loc[:2,'housing_median_age']
```

```python
df_csv.loc[1]
```

# Grouping

```python
df_min = df_csv.groupby('housing_median_age')['median_house_value'].min()
df_max = df_csv.groupby('housing_median_age')['median_house_value'].max()
df_mean = df_csv.groupby('housing_median_age')['median_house_value'].mean()
pd.merge(df_min,df_max, on='housing_median_age')
```
